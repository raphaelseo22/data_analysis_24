````python
import pandas as pd
import numpy as np
df = pd.read_csv('uci-secom.csv')
df.head()
````

````python
df.shape
````

````python
df['Pass/Fail'].value_counts(normalize=True)
````

````python
object_col = []
num_col = []
for c in df.columns:
    if df[c].dtype == 'object':
        object_col.append(c)
    elif (df[c].dtype == 'float64') or (df[c].dtype == 'int64'):
        num_col.append(c)
````

````python
# object 컬럼 1개 나머지는 숫자형 컬럼
print(len(object_col), len(num_col))
````

````python
have_null_col = []
no_null_col = []
for c in df.columns:
    null_sum = df[c].isnull().sum()
    if null_sum == 0:
        no_null_col.append(c)
    else:
        have_null_col.append(c)
````

````python
print(len(no_null_col), len(have_null_col))
````

````python
null_over_700 = []
for c in have_null_col:
    if df[c].isnull().sum() >= 700:
        null_over_700.append(c)
````

````python
len(null_over_700)
````

````python
df2 = df.drop(null_over_700, axis=1)
df2 = df.drop('Pass/Fail', axis=1)
y = df['Pass/Fail']
````

````python
num_columns = df2._get_numeric_data().columns.tolist()
````

````python
ls1, ls2 = [], []
for i in num_columns:
    if len(df2[i].unique()) == 2:
        ls1.append(i)
        if df2[i].isnull().sum() > 0:
            ls2.append(i)

print(len(ls1) == len(ls2))
print(len(ls1))
constants = ls1
````

````python
df3 = df2.drop(constants+ls2, axis=1)
````

````python
df3.shape
````

````python
# 중복이 존재하지 않음
df3[df3.duplicated(keep=False)]
````

````python
def cal_outlier_condition(df, c):
    q1=df[c].quantile(0.25)
    q3=df[c].quantile(0.75)
    iqr=q3-q1
    return (q3 + 1.5 * iqr, q1 - 1.5 * iqr)
````

````python
num_columns = df3._get_numeric_data().columns.tolist()
````

````python
# IQR 기반 아웃라이어 검사
outlier_num = {}
for c in df3[num_columns].columns:
    condition = cal_outlier_condition(df3, c)
    # o_n = len(df2[(df2[c] > condition[0]) | (df2[c] < condition[1])])
    outlier_num[c] = df3[(df3[c] > condition[0]) | (df3[c] < condition[1])]
````

````python
pvalue = []
for c in df_fill_na.columns:
    t, p = stats.ttest_ind(fill_df_y[fill_df_y['Pass/Fail']==-1][c], fill_df_y[fill_df_y['Pass/Fail']==1][c], equal_var=True)
    if p <= 0.05:
        pvalue.append((c, p))
````
