```python
import time
import requests

import pandas as pd
from tqdm import tqdm
from constants import API_KEY
from utils import check_count
```

```python
df = pd.read_csv('lol_user_with_puuid.csv')
```

```python
df.head()
```

```python
from constants import MATCH_ID_URL
```

```python
match_list = []
start = time.time()
cnt = 0
for puuid in tqdm(df['puuid']):
    url = MATCH_ID_URL.format(puuid=puuid, api_key=API_KEY)
    match_id = requests.get(url).json()
    cnt, start = check_count(cnt, start)
    if len(match_id) != 0:
        match_list.extend(match_id)
```

```python
type(requests.get(url).json())
```

```python
len(match_list)
```

```python
unique_match_list = list(set(match_list))
```

```python
MATCH_INFO_URL = 'https://asia.api.riotgames.com/lol/match/v5/matches/{match_id}?api_key={api_key}'
```

```python
match_df = pd.DataFrame(data=unique_match_list, columns=['match_id'])
```

```python
match_df.to_csv('match_id.csv')
```

```python
match_info_list = []
start = time.time()
cnt = 0
for match_id in tqdm(unique_match_list):
    url = MATCH_INFO_URL.format(match_id=match_id, api_key=API_KEY)
    match_info = requests.get(url).json()
    cnt, start = check_count(cnt, start)
    match_info_list.append(match_info)
```
