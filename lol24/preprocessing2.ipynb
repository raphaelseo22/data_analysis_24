```python
def drop_regex(data: pd.DataFrame, pattern: str) -> pd.DataFrame:
    filtered_columns = data.filter(regex=pattern)
    data = data.drop(filtered_columns, axis=1)
    return data
```

```python
pattern = '.*vision.*'
```

```python
vision_columns = data.filter(regex=pattern)
```

```python
pattern = ".*Pings"
```

```python
ping_columns = data.filter(regex=pattern)
```

```python
blue_ping = ping_columns.filter(regex='blue_.*')
red_ping = ping_columns.filter(regex='red_.*')
```

```python
total['blue_ping'] = blue_ping.sum(axis=1)
total['red_ping'] = red_ping.sum(axis=1)
```
